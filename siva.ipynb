```python
import os
import cv2
import numpy as np
import face_recognition as fr
```

```python
path='images'
```

```python
my_list=os.listdir(path)
my_list
```

```python
imgs=[] # to assign array value == read image
classnames=[] # to assign ouput
for img in my_list:
    imgpath=os.path.join(path,img)
    # print(imgpath)
    current_img=cv2.imread(imgpath)
    # print(current_img)
    imgs.append(current_img)
    # classnames.append(img[:,-4])
    classnames.append(img.split('.')[0]) 
print(classnames)
```

```python
# Face Location and Face encoding 
img=imgs[0]
# img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
# face_in_frame=fr.face_locations(img,3)
# print(face_in_frame)  
# # encoding 
# encode=fr.face_encodings(img,face_in_frame)[0]
# print(encode) 
```

```python
encode_list=[]
def get_face_encoding(images):
    for img in images:
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
        face_in_frame=fr.face_locations(img)
        encode=fr.face_encodings(img,face_in_frame)[0]
        encode_list.append(encode)
    return encode_list
print(get_face_encoding(imgs)) # function call    
```

```python
encodelist_knownfaces=get_face_encoding(imgs)
print(encodelist_knownfaces) 
```

```python
video=cv2.VideoCapture(0) # 0 same carea using
while True:
    suc,frame=video.read()
    frame1=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    face_in_frame=fr.face_locations(frame1)
    # print(face_in_frame)
    encoded_face=fr.face_encodings(frame1,face_in_frame)
    # print(encoded_face) 

    for enc_face,faceloc in zip(encoded_face,face_in_frame):
        matches=fr.compare_faces(encodelist_knownfaces,enc_face)
        print(matches) 
        # face distance 
        face_dr=fr.face_distance(encodelist_knownfaces,enc_face)
        print(face_dr)
        
        matchindex=np.argmin(face_dr)
        print(matchindex)

        if matches[matchindex]:
            name=classnames[matchindex]
            print(name)
        else:
            name='unknown' 
            print(name)
        y1,x2,y2,x1=faceloc
        cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),3)
        cv2.putText(frame,name,(x1,y1),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)

    cv2.imshow("Face",frame)
    if cv2.waitKey(1)& 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
